```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,9)
y = x **2
print(x)
print(y)
```

```python
plt.plot(x,y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sample Graph')
plt.show()
plt.hist(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r*-')
plt.subplot(1,2,2)
plt.plot(y,x, 'g+-')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('set title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('set x label')
axes1.set_ylabel('set y label')
axes1.set_title('set 1 title')

axes2.plot(y,x,'r')
axes2.set_xlabel('set x label')
axes2.set_ylabel('set y label')
axes2.set_title('set 2 title')
```

```python
x = np.linspace(0, 10, 100)   # corrected here
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.plot(x, y1, label='Sine')
plt.title('Sine Plot')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='orange', label='Cosine')
plt.title('Cosine Plot')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
x =np.linspace(0,10,100)
plt.plot(x, np.sin(x) , label='Sine',linewidth=2)
plt.plot(x, np.cos(x) , label='Cosine',linestyle='--',linewidth=2)
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)

plt.annotate('Peak', xy=(np.pi/2,1), xytext=(2,0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('Peak', xy=(np.pi/8,1), xytext=(8,0.25),
             arrowprops=dict(facecolor='yellow', arrowstyle='->'))
plt.title('plot with Annotation')
plt.show()
```
